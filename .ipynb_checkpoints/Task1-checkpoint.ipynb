Используя исторические данные на листе «Timeseries» (см. Задания_1_2.xlsx), построить модель временного ряда.  Спрогнозировать  ежедневное поведение ряда в следующие 3 мес.  Пояснить выбор метода прогнозирования. Привести оценки качества прогноза.

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import warnings
import seaborn as sns
from itertools import product
%pylab inline
```

```python
ts = pd.read_excel("Задания_1_2.xlsx", sheet_name = 'Timeseries')
```

```python
#Изучим данные
ts.head()
```

```python
ts.info()
```

```python
#Присвоим индекс полю Дата
ts.index = ts['Date']
#Удалим столбец Даты
ts = pd.DataFrame(ts['series1'])
```

```python
ts.plot(figsize=(14, 8), title="Временной ряд")
```

По данному графику можно сказать, что тренд и сезонность отсутствуют. Сильновыделяющиеся значения возможно могут быть аномалиями. 

```python
#Проверим данные на пропуски
ts.isna().sum()
```

```python
#Посмотрим на характеристики данных
ts.describe()
```

Среднее примерно равно медиане, это говорит о том, что данные тяготеют к среднему, однако макс значения очень большое, необходимо разобраться с аномалиями. А также достаточно больше std по сравнению со средним.

```python
#Построим ящик с усами и посмотрим на сильно выделяющиеся значения и выявим аномалии
plt.figure(figsize=(14,7))
sns.boxplot(x=ts['series1'])
```

Все, что находится за пределми усов диаграммы размаха-это аномалии. Однако в данном случае их слишком много. Удалим сильно выбивающиеся

```python
ts[ts['series1'] < 1000]
```

```python
#Посчитаем критерий Дики-Фуллера
print ("Критерий Дики-Фулера: p=%f" % sm.tsa.stattools.adfuller(ts)[1])
```

Критерий Дики-Фулера меньше 5%, что говорит о том, что ряд можно считать стационарным, такой ряд можно прогнозировать. Воспользуемся методом ARIMA.

```python
plt.figure(figsize=(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(ts.values.squeeze(), lags = 28, ax = ax )
pylab.show()
plt.figure(figsize=(15,8))
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(ts.values.squeeze(), lags = 28, ax = ax )
pylab.show()
```
